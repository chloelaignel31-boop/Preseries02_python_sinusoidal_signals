# Sine VS Cosine (Question 6)

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 1               # Amplitude
f = 1               # Frequency in Hz
omega = 2 * np.pi * f  # Angular frequency
t = np.linspace(0, 2, 500)  # Time from 0 to 2 seconds

# Signals
y_sin = A * np.sin(omega * t)  # Sine wave
y_cos = A * np.cos(omega * t)  # Cosine wave

# Plot
plt.figure(figsize=(8,4))
plt.plot(t, y_sin, label='Sine', color='blue')
plt.plot(t, y_cos, label='Cosine', color='red', linestyle='--')
plt.title('Sine vs Cosine')
plt.xlabel('Time t (s)')
plt.ylabel('Amplitude y(t)')
plt.legend()
plt.grid(True)

# Display in the notebook
plt.show()
```
